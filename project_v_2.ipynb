```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset
file_path = "/content/newds - newds.csv"  # Update the path if necessary
df = pd.read_csv(file_path)
df.drop("Unnamed: 9",axis=1, inplace=True)
# Data Preprocessing

# Convert to numeric types
df['Year of Release'] = pd.to_numeric(df['Year of Release'], errors='coerce')
df['Metascore'] = pd.to_numeric(df['Metascore'], errors='coerce')

# Drop rows with missing target variable ('Gross')
df.dropna(subset=['Gross'], inplace=True)

# One-hot encode categorical variables
df = pd.get_dummies(df, columns=['Genre'], drop_first=True)

# Drop unnecessary columns
df.drop(columns=['Movie Title'], inplace=True)

# Define features and target
X = df.drop(columns=['Gross'])
y = df['Gross']

# Handle missing values
imputer = SimpleImputer(strategy='median')
X_imputed = imputer.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Train Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Evaluate the model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Model Trained - MSE: {mse}, R-squared: {r2}")

def get_user_input():
    print("Enter movie details:")
    year_of_release = float(input("Year of Release: "))
    metascore = float(input("Metascore: "))
    budget = float(input("Budget: "))
    runtime = float(input("Runtime (minutes): "))
    votes = float(input("Votes: "))
    imdb_rating = float(input("IMDB Rating: "))

    # Assuming genre is one-hot encoded, initialize all as 0
    genre_dict = {genre: 0 for genre in X.columns if genre.startswith('Genre_')}
    selected_genre = input(f"Enter genre ({', '.join(genre_dict.keys())}): ")
    if selected_genre in genre_dict:
        genre_dict[selected_genre] = 1

    # Combine all inputs into a feature array
    user_features = [year_of_release, metascore, budget, runtime, votes, imdb_rating] + list(genre_dict.values())
    return np.array(user_features).reshape(1, -1)

# Get user input and predict
data = get_user_input()
data_imputed = imputer.transform(data)
predicted_gross = model.predict(data_imputed)

print(f"Predicted Gross Revenue: ₹ {predicted_gross[0]*0.0000001:,.2f} Crs")
```

```python
import numpy as np
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset
file_path = "/content/newds - newds.csv"  # Update path if necessary
df = pd.read_csv(file_path)
df.drop("Unnamed: 9", axis=1, inplace=True, errors='ignore')

# Convert to numeric types
df['Year of Release'] = pd.to_numeric(df['Year of Release'], errors='coerce')
df['Metascore'] = pd.to_numeric(df['Metascore'], errors='coerce')

# Drop rows with missing target variable ('Gross')
df.dropna(subset=['Gross'], inplace=True)

# Define specific genre columns
genre_columns = ['Genre_Action', 'Genre_Comedy', 'Genre_Drama', 'Genre_Horror', 'Genre_SciFi']
df[genre_columns] = 0  # Initialize all genres as 0

# Assign one-hot encoding for genres based on actual dataset
if 'Genre' in df.columns:
    for index, row in df.iterrows():
        if row['Genre'] in genre_columns:
            df.at[index, f'Genre_{row["Genre"]}'] = 1
    df.drop(columns=['Genre'], inplace=True)

# Drop unnecessary columns
df.drop(columns=['Movie Title'], inplace=True, errors='ignore')

# Define features and target
X = df.drop(columns=['Gross'])
y = df['Gross']

# Handle missing values
imputer = SimpleImputer(strategy='median')
X_imputed = imputer.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Train Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Save model and imputer as pickle files
# Save the model and imputer
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)

with open('imputer.pkl', 'wb') as f:
    pickle.dump(imputer, f)

print("Model and Imputer saved as pickle files.")
```
